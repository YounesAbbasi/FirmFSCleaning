## Cleansing

```python
def cleansingFunction(df,path):
    df = df.iloc[:,:15]
    df.drop(columns=['nemad','company_name', 'reported_date_m'], inplace=True)

    for i in range(0,len(df)):
        if df.iloc[i,2] == df.iloc[i,7]:
            df.iloc[i,3] = df.iloc[i,8]
        else:
            df.iloc[i,3] = df.iloc[i,3]
    
    df.drop(columns=['is_total_value','reported_audit_status'], inplace=True)

    df.reported_period = df.reported_period.apply(lambda x: 4)

    df.date = df.date.apply(lambda x: x.replace('-',''))

    df.reported_date_j = df.reported_date_j.apply(lambda x: x.replace('-',''))

    df.value = df.value.apply(lambda x: str(x).replace(',',''))

    for row in range(0,len(df)):
        if df.iloc[row,1][-1] == ')':
            df.iloc[row,1] = df.iloc[row,1].replace(')','')
            df.iloc[row,1] = df.iloc[row,1].replace('(','')
            df.iloc[row,1] = '-'+df.iloc[row,1]
        else:
            df.iloc[row,1] = df.iloc[row,1]

    df.value = df.value.apply(lambda x: int(x))

    for row in range(0,len(df)):
        if df.iloc[row,3]=='audited':
            df.iloc[row,3] = 1
        elif df.iloc[row,3] == 'not_audited':
            df.iloc[row,3] = 0
        else:
            df.iloc[row,3] = 2
    
    CodeDf = pd.read_excel(path)

    CodeDf = CodeDf[['Title','Coding']]

    df = df.merge(CodeDf, how='left', left_on='description', right_on='Title')

    df.drop(columns=['description','Title'], inplace=True)

    return df
```

```python
BS = cleansingFunction(BSDf, r'D:\Python_project\FirmFSCleansing\processed\FSTitleCodning\BS\BSTitleForCodingAddedCoding_Abbasi_14011026.xlsx')
```

```python
IS = cleansingFunction(PLDf , r'D:\Python_project\FirmFSCleansing\processed\FSTitleCodning\IS\ISTitleForCodingAddedCoding_Abbasi_14011027.xlsx')
```
