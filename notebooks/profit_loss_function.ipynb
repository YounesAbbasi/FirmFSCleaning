```python
def file_path(path):
    if __name__ == "__main__":
        balance_sheet = []
        profit_loss = []
        for (root, dirs, files) in os.walk(path , topdown = True):
            if len(files) == 0:
                continue
            else:
                balance_sheet.append(root + '\\' + files[0])
                profit_loss.append(root + '\\' + files[1])
        return balance_sheet, profit_loss
```
